## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """
    Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')
    """
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """
    Applies the Canny transform
    """
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """
    Applies a Gaussian Noise kernel
    """
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

# Read in img and convert to grayscale
img_init_file = 'test_images/solidWhiteCurve.jpg'
img_init = mpimg.imread(img_init_file)
img_gray = grayscale(img_init)

# Apply Gaussian smoothing (reduce noise)
img_blur = gaussian_blur(img_gray, kernel_size=5)

# Apply Canny edge detection to the image
img_edges = canny(img_blur, low_threshold=50, high_threshold=150)

# Define region of interest (ignore everything else!)
shape = img_init.shape
bottom_left   = (75, shape[0])
top_left  = (400, 325)
top_right = (shape[1]-400, 325)
bottom_right  = (shape[1], shape[0])
vertices = np.array([[bottom_left, top_left, top_right, bottom_right]], dtype=np.int32)
img_edges_roi = region_of_interest(img_edges, vertices=vertices)

# Apply Hough transform (to detect line segments)
rho = 1               # distance resolution (in pixels of the Hough grid)
theta = np.pi/180     # angular resolution  (in radians of the Hough grid)
threshold = 30         # min number of votes (intersections in Hough grid cell)
min_line_len = 20     # min number of pixels making up a line
max_line_gap = 15      # max gap in pixels between connectable line segments
img_lines = hough_lines(img_edges_roi, rho, theta, threshold, min_line_len, max_line_gap)

# Overlay detected line segments on original image
img_final = weighted_img(img_lines, img_init)
plt.imshow(img_edges_roi)
````

## Improve the draw_lines() function

**At this point**, if you were successful with making the pipeline and tuning parameters, you probably have the Hough line segments drawn onto the road, but what about identifying the full extent of the lane and marking it clearly as in the example video ([P1_example.mp4](./examples/P1_example.mp4))?  Think about defining a line to run the full length of the visible lane based on the line segments you identified with the Hough Transform. As mentioned previously, try to average and/or extrapolate the line segments you've detected to map out the full extent of the lane lines. 

**Objective**: Go back and modify your draw_lines function accordingly and try re-running your pipeline. The new output should draw a single, solid line over the left lane line and a single, solid line over the right lane line. The lines should start from the bottom of the image and extend out to the top of the region of interest.

````python
import pandas as pd
def draw_solid_lines(img, lines, y_top, color=[255, 0, 0], thickness=12):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    left_m=[]; left_b=[]; left_x=[]; 
    right_m=[]; right_b=[]; right_x=[]; 
    shape = img.shape
    for line in lines:
        m=[]; b=[]; x=[]; y=[]
        for x1,y1,x2,y2 in line:
            mmm = (y2-y1)/(x2-x1)
            m.append(mmm)
            b.append(np.array((y1-mmm*x1, y2-mmm*x2)).mean())
            x.append(x1); x.append(x2)
            y.append(y1); y.append(y2)
        
        # Find avg slope of the line    
        m_line = np.array(m).mean()
        b_line = np.array(b).mean()
        
        # Is it on the left or right side of the car?
        #   The left lane line will have a negative slope, while the right lane
        #   line will have a positive slope.  (Remember: positive y goes down the
        #   image.) Just as we defined a regional mask,
        #   we might further define an angular mask to help ensure we only look
        #   at lane line samples.   
        #
        #   From looking at the images, the left lane line should have an angle
        #   around 45* (pi/4), while the right lane should have an angle around
        #   -45* (-pi/4). So one way to do this is to define some epsilon that
        #   provides a range around these values, e.g., +/- 10* (i.e., +/- 0.17 rad).
        epsilon = 0.25 # 
        right_range  = ( np.pi/4 - epsilon,  np.pi/4 + epsilon)
        left_range = (-np.pi/4 - epsilon, -np.pi/4 + epsilon)
        if (m_line >= left_range[0]) & (m_line <= left_range[1]):
            left_m.append(m_line)
            left_b.append(b_line)
        elif (m_line >= right_range[0]) & (m_line <= right_range[1]):
            right_m.append(m_line)
            right_b.append(b_line)
    
    # ASSURE LEFT LINE ESTIMATE EXISTS
    #   * y_min gives larget "height" of line
    #   * x_min = (y_min-b)/m
    left_m = np.array(left_m).mean()
    left_b = np.array(left_b).mean()
    if np.isnan([left_m, left_b]).sum() > 0:
        # If for some reason, present frame could not detect lane lines, use
        #    the information from the last frame
        left_m = pd.read_pickle('most_recent_left_m')
        left_b = pd.read_pickle('most_recent_left_b')
    pd.to_pickle(left_m, 'most_recent_left_m')
    pd.to_pickle(left_b, 'most_recent_left_b')
    
    
    # ASSURE RIGHT LINE ESTIMATE EXISTS
    right_m = np.array(right_m).mean()
    right_b = np.array(right_b).mean()
    if np.isnan([right_m, right_b]).sum() > 0:
        # If for some reason, present frame could not detect lane lines, use
        #    the information from the last frame
        right_m = pd.read_pickle('most_recent_right_m')
        right_b = pd.read_pickle('most_recent_right_b')
    pd.to_pickle(right_m, 'most_recent_right_m')
    pd.to_pickle(right_b, 'most_recent_right_b')
    
    # Line Bottoms
    left_ybot = shape[1]  
    left_xbot = int((left_ybot - left_b)/ left_m)
    right_ybot = shape[1]
    right_xbot = int( (right_ybot - right_b)/right_m)
    
    # Line Tops
    right_xtop = int((y_top - right_b) / right_m)
    left_xtop = int((y_top - left_b) / left_m)
    
    # Draw Lines
    cv2.line(img, (left_xbot, left_ybot), (left_xtop, y_top), color, thickness)
    cv2.line(img, (right_xtop, y_top), (right_xbot, right_ybot), color, thickness)    
    
            
def hough_solid_lines(img, rho, theta, threshold, min_line_len, max_line_gap, y_top):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(
        img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, 
        maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_solid_lines(line_img, lines, y_top)
    return line_img


def process_image2(
    image, 
    kernel_size=5,
    edge_low = 50,
    edge_high = 150,
    rho = 1,
    theta = np.pi/180,
    threshold=30,
    min_line_len = 20,
    max_line_gap = 15,
    hough_fcn = hough_solid_lines
):   
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    #img_init = mpimg.imread(image)
    
    img_h = image.shape[0]
    img_w = image.shape[1]
    roi_from_top = int(0.63*img_h)
    roi_bot_from_side = int(0.1*img_w)
    roi_top_from_side = int(0.40*img_w)
    
    roi_bl=(roi_bot_from_side, img_h)
    roi_tl=(roi_top_from_side, roi_from_top)
    roi_tr=(img_w - roi_top_from_side, roi_from_top)
    roi_br=(img_w - roi_bot_from_side, img_h)
    
    img_gray = grayscale(image)

    # Apply Gaussian smoothing (reduce noise)
    img_blur = gaussian_blur(img_gray, kernel_size=kernel_size)

    # Apply Canny edge detection to the image
    img_edges = canny(img_blur, low_threshold=edge_low, high_threshold=edge_high)

    # Define region of interest (ignore everything else!)
    vertices = np.array([[roi_bl, roi_tl, roi_tr, roi_br]], dtype=np.int32)
    img_edges_roi = region_of_interest(img_edges, vertices=vertices)

    # Apply Hough transform (to detect line segments)
    img_lines = hough_fcn(img_edges_roi, rho, theta, threshold, min_line_len, max_line_gap, roi_from_top)

    # Overlay detected line segments on original image
    img_final = weighted_img(img_lines, image)
    return img_final
````
